```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import platypus
```

```python
from platypus import NSGAII, Problem, Real

def area_calc(vars):
    return vars[0]*vars[1]-(vars[3]-1)*(vars[0]-2),vars[2]*vars[3]

problem = Problem(4,2)   # x1,y1,x2,y2
problem.types[:] = [Real(0, 3),Real(1,4),Real(1,4),Real(0,3)]
problem.directions[:] = Problem.MAXIMIZE
problem.function = area_calc

algorithm = NSGAII(problem,population_size=30)
algorithm.run(10000)
```

```python
algorithm.evaluate_all(algorithm.result)
```

```python
plt.scatter([s.objectives[0] for s in algorithm.result],
            [s.objectives[1] for s in algorithm.result])
plt.xlabel('Area 2')
plt.ylabel('Area 1')
```

```python
from matplotlib.patches import Rectangle
```

```python
n = len(algorithm.result)
for i in range(0,n):
    res = algorithm.result[i]
    plt.figure(figsize=(4,4))
    area = res.objectives[0]+res.objectives[1]
    plt.title('Solution '+str(i)+' area: '+str(area))
    ax = plt.gca()
    ax.add_patch(Rectangle((0,0),height=4,width=4,fill=None))
    ax.add_patch(Rectangle((0,0.25),res.variables[0]*0.25,res.variables[1]*0.25))
    ax.add_patch(Rectangle((0.25,0),res.variables[2]*0.25,res.variables[3]*0.25,color='red'))
    ax.add_patch(Rectangle((0.25,0.25),(res.variables[0]-2)*0.25,(res.variables[3]-1)*0.25))
```
